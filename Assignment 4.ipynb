<h2> Adventure Game : Haunted House <h2>

```python
#Game plot around a different block and area of haunted house 
haunted_house = {'entrance': 'Sssshhhhh!!!! You are enter in haunted house, unlock the door, where the dark yellow light is flashing',
                 'mainhall' : 'Now you enter in big main hall, on top of you broken chandeliers and front of broken stairs',
                 'kitchen': 'You enter in kitchen, you see on the table beheaded body with spillage of blood on the floor',
                 'library': 'You arrived in library and found some horror mystery books and one the book give you direction to out of this house',
                 'bedroom': 'In bedroom you found many two human skeleton talking and playing each to other, NOW YOU JUST RUUUUUUNNNNN',
                 'study_room':'You are at study room and see something strange that beheaded body typing on python program',
                 'bathroom':'You just open a bathroom door and found a dead body bathing with blood under the shower',
                 'escape_door':'Oooooohhhhh, Thank God, Finally you come out!!!!'}

# the map has describe which area you can access with direction like north,south,east,wast
map = {'entrance':{'north':'mainhall'},
       'mainhall':{'north':'library','south':'entrance','east':'bedroom','wast':'kitchen'},
       'kitchen':{'north':'escape_room','wast':'mainhall'},
       'library':{'north':'study_room','south':'mainhall'},
       'bedroom':{'north':'bathroom','east':'mainhall'},
       'study_room':{'south':'library','east':'escape_door','wast':'bathroom'},
       'bathroom':{'south':'bathroom','east':'study_room'},
       'escape_door':{'south':'kitchen','wast':'study_room'}}

# A current room variable has show user present location 
current_room = 'entrance'
# System has count to user activity during game
count_step = 0
print('WELCOME to HAUNTED HOUSE')

while True:
    # This statement print a location of user before start a new loop
    print(haunted_house[current_room])
    # User has option to add their input for direction
    journey = input('please select your way : north,south,east,wast or quit to exit: ').lower()
    # Input has option to exit game and it's break a loop
    if journey == 'quit':
        print('Thanks for playing game of haunted house')
        break
    # This loop has access different location, direction and description 
    if journey in map[current_room].keys():
        current_room = map[current_room][journey]
        count_step += 1 
        print(f'You are waking {count_step} step in this house.')
    # This house has define boundary 
    else:
        print('You can not go this way, There is a wall!!')
```
